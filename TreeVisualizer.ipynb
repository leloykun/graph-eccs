```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Load the Trees

```python
# Tree size; replace this if you want to visualize other sizes
N = 9
```

```python
graphs = []
# If you're opening this on Google Colab, you first have to
# upload the tree data first. Click the "Files" tab on the left
with open("trees/tree{}.txt".format(N), 'r') as f:
    M = int(f.readline())
    for _ in range(M):
        _ = int(f.readline())
        G = nx.Graph()
        edges = []
        for _ in range(N-1):
            u, v = map(int, f.readline().split())
            edges.append((u, v))
        G.add_edges_from(edges)
        graphs.append(G)
```

```python
print("We have {} tree(s) of size {}".format(M, N))
```

### Draw the Trees

```python
nx.draw(graphs[0])
```
